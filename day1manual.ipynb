```python
import numpy as np
import math as math
import matplotlib.pyplot as plt
```

```python
def inverse_pdf(x):
    return 100*math.sqrt(x)+900
array = np.zeros((100, 100))

for i in range(1000):
    turtles = np.empty(1000)
    for j in range(1000):
        num = np.random.uniform(0,1)
        turtles[j] = inverse_pdf(num)
    for j in range(100):
        for k in range(j, 100):
            for l in range(1000):
                if turtles[l] > 900 + k:
                    continue
                if turtles[l] < 900 + k and turtles[l] > 900 + j:
                    array[j][k] = array[j][k] + 1000 - (900 + k)
                else:
                    array[j][k] = array[j][k] + 1000 - (900 + j)
print(array / 1000)
```

```python
plt.imshow(array, cmap='hot', interpolation='nearest')
plt.colorbar()  # Add a color bar legend
plt.show()
```
